```python
import pandas as pd

data = {
    'Vin': [12, 12, 24, 24, 48, 48, 12, 24, 48, 12],
    'Iout': [2.0, 3.5, 1.8, 3.0, 1.2, 2.5, 4.0, 4.5, 3.0, 2.5],
    'Temp': [45, 68, 55, 72, 60, 85, 80, 90, 78, 50],
    'Efficiency': [96.5, 93.2, 95.8, 91.4, 94.1, 88.7, 89.3, 86.1, 90.2, 96.0]
}

df = pd.DataFrame(data)
df.to_csv('converter_data.csv', index=False)
print("فایل converter_data.csv ساخته شد!")
print(df)
```

```python
# --- نصب کتابخانه‌ها ---
!pip install scikit-learn matplotlib > /dev/null 2>&1  # بدون چاپ خروجی

# --- لود داده ---
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

df = pd.read_csv('converter_data.csv')
print("Converter Data:")
print(df)

# --- مدل ML ---
X = df[['Vin', 'Iout', 'Temp']]
y = df['Efficiency']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# --- دقت ---
pred = model.predict(X_test)
mae = mean_absolute_error(y_test, pred)
print(f"\nModel Accuracy (MAE): {mae:.2f}%")

# --- پیش‌بینی جدید ---
new_input = np.array([[24, 3.5, 80]])
pred_eff = model.predict(new_input)[0]
status = "Risk of Drop" if pred_eff < 90 else "Excellent"
print(f"Predicted Efficiency at 80°C: {pred_eff:.1f}% → {status}")

# --- نمودار ---
plt.figure(figsize=(9,6))
scatter = plt.scatter(df['Temp'], df['Efficiency'], c=df['Iout'], cmap='plasma', s=100, alpha=0.8)
plt.colorbar(scatter, label='Output Current (A)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Efficiency (%)')
plt.title(f'Buck-Boost Converter Efficiency Prediction\nAccuracy: MAE = {mae:.2f}%')
plt.grid(True, alpha=0.3)
plt.plot(80, pred_eff, 'r*', markersize=15, label=f'Prediction: {pred_eff:.1f}%')
plt.legend()
plt.show()
```
